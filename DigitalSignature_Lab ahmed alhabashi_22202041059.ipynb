```python
# 1. Import necessary Python libraries
import hashlib
from ecdsa import SigningKey, SECP256k1

# -----------------------------------------------
# 2. Generate ECDSA privateâ€“public key pair
# -----------------------------------------------
private_key = SigningKey.generate(curve=SECP256k1)
public_key = private_key.get_verifying_key()

# -----------------------------------------------
# 3. Transaction message + SHA-256 hashing
# -----------------------------------------------
transaction_message = "Ahmed pays 100 coins to Ali"
message_bytes = transaction_message.encode('utf-8')

hash_sha256 = hashlib.sha256(message_bytes).hexdigest()

# -----------------------------------------------
# 4. Sign the hash using the private key
# -----------------------------------------------
signature = private_key.sign(message_bytes)

# -----------------------------------------------
# 5. Verify the signature using the public key
# -----------------------------------------------
is_verified = public_key.verify(signature, message_bytes)

# -----------------------------------------------
# 6. Print results
# -----------------------------------------------
print("Transaction Message:", transaction_message)
print("SHA-256 Hash:", hash_sha256)
print("Signature:", signature.hex())
print("Signature Verified:", is_verified)
```
