```python
import random

# دالة مساعدة لتوليد مفتاح عشوائي
def random_label():
    return random.getrandbits(8)  # 8-bit label for simplicity

# توليد ملصقات لكل طرف
# Omar (A) و Nancy (B)
A_labels = {0: random_label(), 1: random_label()}
B_labels = {0: random_label(), 1: random_label()}

# توليد ملصقات الناتج
# Output labels for AND
OUT_labels = {0: random_label(), 1: random_label()}

# "تشويش" جدول الـ AND
# نستخدم XOR كمثال على تشفير بسيط
garbled_table = {}
for a_val in [0,1]:
    for b_val in [0,1]:
        output_val = a_val & b_val
        # التشفير: XOR ملصق A و B مع ملصق الناتج
        cipher = A_labels[a_val] ^ B_labels[b_val] ^ OUT_labels[output_val]
        garbled_table[(A_labels[a_val], B_labels[b_val])] = cipher

# --------- التنفيذ ---------
# نفترض أن Omar يملك مدخله
Omar_input = 1  # مثلا A=1
Nancy_input = 0 # مثلا B=0

# كل طرف يعطي ملصقه الخاص
A_label = A_labels[Omar_input]
B_label = B_labels[Nancy_input]

# استخدام الجدول المشفر لإيجاد الناتج
cipher = garbled_table[(A_label, B_label)]
reconstructed_output_label = A_label ^ B_label ^ cipher

# التعرف على قيمة الناتج
output_val = None
for val, label in OUT_labels.items():
    if label == reconstructed_output_label:
        output_val = val
        break

print("مدخلات Omar و Nancy:", Omar_input, Nancy_input)
print("الناتج A AND B =", output_val)
```
